##  4.1 Build the Deep learning Model:

```python
from tensorflow.keras.models import Sequential  
from tensorflow.keras.layers import Conv2D,MaxPooling2D,Dense,Flatten,Dropout
```

```python
 there are two specific model buliding API 1.Sequential  1 data input and one output  top to bottom  
    2. functional API
```

```python
then we add layers  1.conv2d-- spacial convolution over images
2. maxpooling2d--> condensing layer 
3.dense  --> connected layer
4. Flatten --> to convert to another format
```

```python
model=Sequential()
```

```python
model.add(Conv2D(16,(3,3),1,activation='relu',input_shape=(256,256,3)))
model.add(MaxPooling2D())
# 1st layer is an input layer having 16 filters,3x3 size filter,stride by 1, activation relu means passing through the function 
# relu-->  negative to 0 positive unchanged
# sigmoid--> one more function
# 256x256 pixels by 3 channels wide

# maxpooling--> is going to convert image data i.e by default it is going to reduce 2x2 and condense the info



model.add(Conv2D(32,(3,3),1,activation='relu'))
model.add(MaxPooling2D())



model.add(Conv2D(16,(3,3),1,activation='relu'))
model.add(MaxPooling2D())


model.add(Flatten())
# condense the single value

model.add(Dense(256,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
# 256 value as output
# and finally one value as output          
          











```

```python
# to compile the model we build
```

```python
model.compile('adam',loss=tf.losses.BinaryCrossentropy(),metrics=['accuracy'])


# adam-->optimizers
# metric is accuracy
```

```python
model.summary()
```
